## Feature Engineering ##
Step 1: We need to change the columnar format of subscription and payments table into row format for analysis

```python
# Separate xactions and amount columns
xactions_cols = [col for col in payments_raw.columns if 'payments-xactions' in col]
amount_cols = [col for col in payments_raw.columns if 'payments-amount' in col]
```

```python
# Create two separate dataframes
payments_xactions_raw = payments_raw[['ID'] + xactions_cols]
payments_amounts_raw = payments_raw[['ID'] + amount_cols]
```

```python
# Melt (unpivot) them
payments_xactions = payments_xactions_raw.melt(
    id_vars='ID',
    var_name='month',
    value_name='payments-xactions'
)

payments_amounts = payments_amounts_raw.melt(
    id_vars='ID',
    var_name='month',
    value_name='payments-amount'
)

subscriptions = subscriptions_raw.melt(
    id_vars='ID',
    var_name='month',
    value_name='subscription_revenue'
)

display(payments_xactions.head())
display(payments_amounts.head())
display(subscriptions.head())
```

```python
# To prove the melt has worked as expected, we are checking if each ID occurs 24 times in payment_mounts table and payment_xactions table
print("payments_amounts summary statistics:", payments_amounts['ID'].value_counts().describe())
print("------------------------------------------------------")
print("payments_xactions summary statistics:", payments_amounts['ID'].value_counts().describe())
print("------------------------------------------------------")
print("subscriptions summary statistics:", subscriptions['ID'].value_counts().describe())
```

```python
# Clean month names
payments_xactions['month'] = payments_xactions['month'].str.extract(r'(\d{4}-\d{2})')
payments_amounts['month'] = payments_amounts['month'].str.extract(r'(\d{4}-\d{2})')
subscriptions['month'] = subscriptions['month'].str.extract(r'(\d{4}-\d{2})')
```

```python
# Merge
payments = pd.merge(
    payments_amounts,
    payments_xactions,
    on=['ID', 'month'],
    how='inner'
)

# Preview
print(payments.head())
print(subscriptions.head())
print("------------------------------------------------------")
print("payments table size: ",len(payments)) # Aligns to 3567*24 = 85608
print("subscription table size: ",len(subscriptions)) # Aligns to 5000*24 = 120000
```

```python
# Clean column names
def clean_column_names(df):
    df.columns = (
        df.columns.str.strip()
                 .str.lower()
                 .str.replace(' ', '_')
                 .str.replace('-', '_')
    )
    return df

# Apply it to all your dataframes
accounts = clean_column_names(accounts)
payments = clean_column_names(payments)
subscriptions = clean_column_names(subscriptions)

# Preview to confirm
print("Accounts columns:", accounts.columns.tolist())
print("Payments columns:", payments.columns.tolist())
print("Subscriptions columns:", subscriptions.columns.tolist())
```

```python
query_temporal_plan_attach = """
SELECT
  strftime('%Y-%m', account_created_date) AS cohort_month,
  subscription_plan,
  COUNT(CASE WHEN payments_enrolled_date IS NOT NULL THEN 1 END) * 1.0 / COUNT(*) AS attach_rate,
  COUNT(*) AS total_accounts
FROM accounts
GROUP BY cohort_month, subscription_plan
ORDER BY cohort_month, subscription_plan
"""

temporal_plan_attach = pysqldf(query_temporal_plan_attach)
```

```python
query_arpa = """
SELECT
    s.month,
    a.subscription_plan,
    a.country,
    a.primary_discipline_bucketed,
    a.licenses,
    SUM(s.subscription_revenue) as subscription_revenue,
    COUNT(DISTINCT case when s.subscription_revenue>0 then s.id end) as account_id_count,
    SUM(DISTINCT case when s.subscription_revenue > 0 then a.licenses end) as licenses_count,
    ROUND(SUM(s.subscription_revenue) * 1.0 / COUNT(DISTINCT s.id),2) AS arpa
FROM subscriptions s
LEFT JOIN accounts a
ON s.id=a.id
GROUP BY
    s.month,
    a.subscription_plan,
    a.country,
    a.primary_discipline_bucketed,
    a.licenses
ORDER BY
    s.month,
    a.subscription_plan,
    a.country,
    a.primary_discipline_bucketed,
    a.licenses
"""
arpa_df = pysqldf(query_arpa)
display(arpa_df)
arpa_df.to_csv('arpa.csv',sep=',',encoding='utf-8')
```

```python
query_churn = """
SELECT
    strftime('%Y-%m', a.account_cancelled_date) AS churn_month,
    a.country,
    a.subscription_plan,
    a.licenses,
    a.primary_discipline_bucketed,
    COUNT(DISTINCT a.id) AS churned_accounts
FROM accounts a
WHERE a.account_cancelled_date IS NOT NULL
GROUP BY churn_month, a.country, a.subscription_plan, a.licenses, a.primary_discipline_bucketed
ORDER BY churn_month, a.country, a.subscription_plan, a.licenses, a.primary_discipline_bucketed
"""
churn_df = pysqldf(query_churn)
display(churn_df)
churn_df.to_csv('account_churn.csv',sep=',')
```

```python
query_payments_activity = """
SELECT
    p.month,
    a.country,
    a.subscription_plan,
    a.licenses,
    a.primary_discipline_bucketed,
    SUM(p.payments_xactions) * 1.0 as payments_xactions,
    SUM(p.payments_amount) * 1.0 as payments_amount,
    COUNT(DISTINCT case when p.payments_xactions > 0 then p.id end) as account_counts_xactions,
    COUNT(DISTINCT case when p.payments_amount > 0 then p.id end) as account_counts_payments,
    SUM(p.payments_xactions) * 1.0 / COUNT(DISTINCT p.id) AS avg_transactions_per_account,
    SUM(p.payments_amount) * 1.0 / COUNT(DISTINCT p.id) AS avg_volume_per_account
FROM payments p
LEFT JOIN accounts a
    ON p.id = a.id
GROUP BY
    p.month,
    a.country,
    a.subscription_plan,
    a.licenses,
    a.primary_discipline_bucketed
ORDER BY
    p.month,
    a.country,
    a.subscription_plan,
    a.licenses,
    a.primary_discipline_bucketed
"""
payments_activity_df = pysqldf(query_payments_activity)
display(payments_activity_df)
payments_activity_df.to_csv('payments_activity.csv',sep=',',encoding='utf-8')
```

```python
subscription_churn_sql = """
WITH monthly_rev AS (
    SELECT
        id,
        month,
        subscription_revenue * 1.0 as subscription_revenue,
        (lag(subscription_revenue) over(partition by id order by month)) * 1.0 as previous_subscription_revenue
    FROM subscriptions
    GROUP BY id, month
)

SELECT  month,
        country,
        subscription_plan,
        licenses,
        primary_discipline_bucketed,
        count(distinct case when subscription_churn_type in ('Subscription Downgrade') then id end) as subscription_downgrade_count,
        sum(distinct case when subscription_churn_type in ('Subscription Downgrade') then (subscription_revenue - previous_subscription_revenue) end) * 1.0 as subscription_downgrade_amount,
        count(distinct case when subscription_churn_type in ('Subscription Churn') then id end) as subscription_churn_count,
        sum(distinct case when subscription_churn_type in ('Subscription Churn') then (subscription_revenue - previous_subscription_revenue) end) * 1.0 as subscription_churn_amount
FROM
(
    SELECT  s.id,
            s.month,
            a.country,
            a.subscription_plan,
            a.licenses,
            a.primary_discipline_bucketed,
            s.subscription_revenue,
            s.previous_subscription_revenue,
            case    when (s.previous_subscription_revenue is NULL or s.previous_subscription_revenue = 0) and s.subscription_revenue > 0 then 'Subscription Begin'
                    when (s.previous_subscription_revenue > 0 and s.subscription_revenue > 0) and s.previous_subscription_revenue = s.subscription_revenue then 'Subscription Continued'
                    when (s.previous_subscription_revenue > 0 and s.subscription_revenue > 0) and s.previous_subscription_revenue > s.subscription_revenue then 'Subscription Downgrade'
                    when (s.previous_subscription_revenue > 0 and s.subscription_revenue > 0) and s.previous_subscription_revenue < s.subscription_revenue then 'Subscription Upgrade'
                    when s.previous_subscription_revenue > 0 and (s.subscription_revenue is NULL or s.subscription_revenue = 0) then 'Subscription Churn'
                    else 'Other'
            end subscription_churn_type
    FROM monthly_rev s LEFT JOIN accounts a
    ON s.id=a.id
)
GROUP BY month, country, subscription_plan, licenses, primary_discipline_bucketed
ORDER BY month, country, subscription_plan, licenses, primary_discipline_bucketed
"""
subscription_churn = pysqldf(subscription_churn_sql)
display(subscription_churn)
subscription_churn.to_csv('subscription_churn.csv',sep=',',encoding='utf-8')
```

```python
payment_churn_sql = """
WITH monthly_payments AS (
    SELECT
        id,
        month,
        payments_xactions,
        lag(payments_xactions) over(partition by id order by month) as previous_payments_xactions,
        lead(payments_xactions) over(partition by id order by month) as next_payments_xactions,
        payments_amount * 1.0 as payments_amount,
        (lag(payments_amount) over(partition by id order by month)) * 1.0 as previous_payments_amount,
        (lag(payments_amount,2) over(partition by id order by month)) * 1.0 as previous_previous_payments_amount,
        (lead(payments_amount) over(partition by id order by month)) * 1.0 as next_payments_amount

    FROM payments
    GROUP BY id, month
)

SELECT  month,
        country,
        subscription_plan,
        licenses,
        primary_discipline_bucketed,
        count(distinct case when payments_churn_type in ('Payment Churn') then id end) as payments_churn_account_count,
        sum(distinct case when payments_churn_type in ('Payment Churn') then (payments_amount - previous_payments_amount) end) * 1.0 as payments_churn_amount
FROM
(
    SELECT  p.id,
            p.month,
            a.country,
            a.subscription_plan,
            a.licenses,
            a.primary_discipline_bucketed,
            p.payments_xactions,
            p.previous_payments_xactions,
            p.next_payments_xactions,
            p.payments_amount,
            p.previous_payments_amount,
            p.next_payments_amount,
            case    when (p.previous_previous_payments_amount is NULL or p.previous_previous_payments_amount = 0)
                            and (p.previous_payments_amount is NULL or p.previous_payments_amount = 0) and p.payments_amount > 0 then 'Payments Begin'
                    when (p.previous_payments_amount > 0 and p.payments_amount > 0 and p.next_payments_amount > 0) then 'Payments Continue'
                    when p.previous_payments_amount > 0 and ((p.payments_amount is null or p.payments_amount = 0) and (p.next_payments_amount is null or p.next_payments_amount = 0)) then 'Payment Churn'
                    else 'Other'
            end as payments_churn_type
    FROM monthly_payments p LEFT JOIN accounts a
    ON p.id=a.id
)
GROUP BY month, country, subscription_plan, licenses, primary_discipline_bucketed
ORDER BY month, country, subscription_plan, licenses, primary_discipline_bucketed
"""
payment_churn = pysqldf(payment_churn_sql)
display(payment_churn)
payment_churn.to_csv('payment_churn.csv',sep=',',encoding='utf-8')
```

```python
account_churn_sql = """
SELECT
    a.id,
    strftime('%Y-%m', a.account_cancelled_date) AS churn_month,
    a.country,
    a.subscription_plan,
    a.licenses,
    a.primary_discipline_bucketed,
    COUNT(DISTINCT a.id) AS churned_accounts
FROM accounts a
WHERE a.account_cancelled_date IS NOT NULL
GROUP BY id, churn_month, a.country, a.subscription_plan, a.licenses, a.primary_discipline_bucketed
ORDER BY id, churn_month, a.country, a.subscription_plan, a.licenses, a.primary_discipline_bucketed
"""
account_churn = pysqldf(account_churn_sql)
display(account_churn)
```

```python
subscription_churn_sql = """
WITH monthly_rev AS (
    SELECT
        id,
        month,
        subscription_revenue * 1.0 as subscription_revenue,
        (lag(subscription_revenue) over(partition by id order by month)) * 1.0 as previous_subscription_revenue
    FROM subscriptions
    GROUP BY id, month
)

SELECT      s.id,
            s.month,
            a.country,
            a.subscription_plan,
            a.licenses,
            a.primary_discipline_bucketed,
            s.subscription_revenue,
            s.previous_subscription_revenue,
            case    when (s.previous_subscription_revenue is NULL or s.previous_subscription_revenue = 0) and s.subscription_revenue > 0 then 'Subscription Begin'
                    when (s.previous_subscription_revenue > 0 and s.subscription_revenue > 0) and s.previous_subscription_revenue = s.subscription_revenue then 'Subscription Continued'
                    when (s.previous_subscription_revenue > 0 and s.subscription_revenue > 0) and s.previous_subscription_revenue > s.subscription_revenue then 'Subscription Downgrade'
                    when (s.previous_subscription_revenue > 0 and s.subscription_revenue > 0) and s.previous_subscription_revenue < s.subscription_revenue then 'Subscription Upgrade'
                    when s.previous_subscription_revenue > 0 and (s.subscription_revenue is NULL or s.subscription_revenue = 0) then 'Subscription Churn'
                    else 'Other'
            end subscription_churn_type
    FROM monthly_rev s LEFT JOIN accounts a
    ON s.id=a.id
"""
subscription_churn = pysqldf(subscription_churn_sql)
display(subscription_churn)
```

```python
payment_churn_sql = """
WITH monthly_payments AS (
    SELECT
        id,
        month,
        payments_xactions,
        lag(payments_xactions) over(partition by id order by month) as previous_payments_xactions,
        lead(payments_xactions) over(partition by id order by month) as next_payments_xactions,
        payments_amount * 1.0 as payments_amount,
        (lag(payments_amount) over(partition by id order by month)) * 1.0 as previous_payments_amount,
        (lag(payments_amount,2) over(partition by id order by month)) * 1.0 as previous_previous_payments_amount,
        (lead(payments_amount) over(partition by id order by month)) * 1.0 as next_payments_amount

    FROM payments
    GROUP BY id, month
)

SELECT  p.id,
            p.month,
            a.country,
            a.subscription_plan,
            a.licenses,
            a.primary_discipline_bucketed,
            p.payments_xactions,
            p.previous_payments_xactions,
            p.next_payments_xactions,
            p.payments_amount,
            p.previous_payments_amount,
            p.next_payments_amount,
            case    when (p.previous_previous_payments_amount is NULL or p.previous_previous_payments_amount = 0)
                            and (p.previous_payments_amount is NULL or p.previous_payments_amount = 0) and p.payments_amount > 0 then 'Payments Begin'
                    when (p.previous_payments_amount > 0 and p.payments_amount > 0 and p.next_payments_amount > 0) then 'Payments Continue'
                    when p.previous_payments_amount > 0 and ((p.payments_amount is null or p.payments_amount = 0) and (p.next_payments_amount is null or p.next_payments_amount = 0)) then 'Payment Churn'
                    else 'Other'
            end as payments_churn_type
    FROM monthly_payments p LEFT JOIN accounts a
    ON p.id=a.id
"""
payment_churn = pysqldf(payment_churn_sql)
display(payment_churn)
payment_churn.to_csv('payment_churn.csv',sep=',')
```

```python
# Account churn already at account_id level with churn month
account_churn = account_churn.rename(columns={'churn_month': 'account_churn_month'})
account_churn = account_churn[['id','account_churn_month']]
display(account_churn)
```

```python
# Collapse subscription churn: Earliest month where churn flag triggers
subscription_churn = subscription_churn[subscription_churn['subscription_churn_type'] == 'Subscription Churn']
subscription_churn = subscription_churn.groupby('id')['month'].min().reset_index().rename(columns={'month': 'subscription_churn_month'})
display(subscription_churn)
```

```python
# Collapse payments churn: Earliest month where payment churn flag triggers (only if 3+ months no payments)
payment_churn = payment_churn[payment_churn['payments_churn_type'] == 'Payment Churn']
payment_churn = payment_churn.groupby('id')['month'].min().reset_index().rename(columns={'month': 'payments_churn_month'})
display(payment_churn)
```

```python
# Merge Churn Dates into Master Table
churn_master = accounts[['id', 'account_created_date', 'country', 'subscription_plan', 'licenses', 'primary_discipline_bucketed']]
churn_master = churn_master.merge(account_churn, on='id', how='left')
churn_master = churn_master.merge(subscription_churn, on='id', how='left')
churn_master = churn_master.merge(payment_churn, on='id', how='left')
display(churn_master)
```

```python
# Define Final Churn Month

def find_final_churn(row):
    churn_dates = []
    if pd.notnull(row['account_churn_month']):
        churn_dates.append(row['account_churn_month'])
    if pd.notnull(row['subscription_churn_month']):
        churn_dates.append(row['subscription_churn_month'])
    return min(churn_dates) if churn_dates else np.nan

churn_master['final_churn_month'] = churn_master.apply(find_final_churn, axis=1)
display(churn_master)
```

```python
# Assumptions
gross_margin = 0.8 # Margin that Jane retains
per_payment_revenue = 0.025 # Jane gets 2.5% of each transaction as revenue
CAC = 400 # Assuming $400 flat across geographies, subscription plans, etc.

subscriptions['subscription_revenue'] = subscriptions['subscription_revenue'] * gross_margin
payments['payments_amount'] = payments['payments_amount'] * per_payment_revenue * gross_margin

subscriptions = subscriptions.merge(churn_master[['id', 'final_churn_month']], on='id', how='left')
payments = payments.merge(churn_master[['id', 'final_churn_month']], on='id', how='left')

subscriptions['month'] = pd.to_datetime(subscriptions['month'])
payments['month'] = pd.to_datetime(payments['month'])
subscriptions['final_churn_month'] = pd.to_datetime(subscriptions['final_churn_month'])
payments['final_churn_month'] = pd.to_datetime(payments['final_churn_month'])

subscriptions = subscriptions[subscriptions['month'] <= subscriptions['final_churn_month']]
payments = payments[payments['month'] <= payments['final_churn_month']]

sub_revenue = subscriptions.groupby('id')['subscription_revenue'].sum().reset_index()
pay_revenue = payments.groupby('id')['payments_amount'].sum().reset_index()
revenue_data = sub_revenue.merge(pay_revenue, on='id', how='outer').fillna(0)
revenue_data['total_revenue'] = revenue_data['subscription_revenue'] + revenue_data['payments_amount']
```

```python
# Step 8: LTV, & Payback
final_metrics = churn_master.merge(revenue_data, on='id', how='left')
final_metrics['ltv'] = final_metrics['total_revenue']
final_metrics['cac'] = CAC

final_metrics['final_churn_month'] = pd.to_datetime(final_metrics['final_churn_month'])

final_metrics['avg_monthly_revenue'] = final_metrics['ltv'] / final_metrics['final_churn_month'].apply(lambda x: (x.year - pd.to_datetime('2023-01-01').year)*12 + (x.month - 1) if pd.notnull(x) else np.nan)
final_metrics['payback_period_months'] = final_metrics['cac'] / final_metrics['avg_monthly_revenue']
```

```python
final_output = final_metrics[['id', 'account_created_date', 'country', 'subscription_plan', 'licenses', 'primary_discipline_bucketed',
                              'final_churn_month', 'ltv', 'cac', 'payback_period_months']]

display(final_output)
```

```python
# Fill Missing LTVs Based on Exact Segment including Account Created Month-Year

# Extract account_created_month_year
final_output['account_created_month_year'] = pd.to_datetime(final_output['account_created_date']).dt.to_period('M').astype(str)

# Identify where LTV is missing
ltv_missing = final_output['ltv'].isnull() | (final_output['ltv'] == 0)

# Calculate median LTV for each segment (Country + Subscription Plan + Licenses + Primary Discipline + Account Created Month-Year)
segment_medians = final_output[~ltv_missing].groupby(
    ['country', 'subscription_plan', 'licenses', 'primary_discipline_bucketed', 'account_created_month_year']
)['ltv'].median().reset_index().rename(columns={'ltv': 'segment_median_ltv'})

# Merge back segment median
final_output = final_output.merge(segment_medians, on=['country', 'subscription_plan', 'licenses', 'primary_discipline_bucketed', 'account_created_month_year'], how='left')

# Fill missing LTVs
final_output['ltv'] = np.where(ltv_missing, final_output['segment_median_ltv'], final_output['ltv'])

# Recalculate Payback Period
final_output['avg_monthly_revenue'] = final_output['ltv'] / (
    final_output['final_churn_month'].apply(lambda x: (x.year - pd.to_datetime('2023-01-01').year)*12 + (x.month - 1) if pd.notnull(x) else np.nan)
)
final_output['payback_period_months'] = final_output['cac'] / final_output['avg_monthly_revenue']

# Drop helper median column
final_output.drop(columns=['segment_median_ltv'], inplace=True)

# Final Preview
final_output.head()
```
